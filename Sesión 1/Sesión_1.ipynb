# Preludio. Brevísima historia de la programación

![Imagen](https://upload.wikimedia.org/wikipedia/commons/thumb/8/80/Arts_et_Metiers_Pascaline_dsc03869.jpg/1024px-Arts_et_Metiers_Pascaline_dsc03869.jpg)

En 1642 Blaise Pascal inventa la primera calculadora que funciona con ruedas y engranajes.

1801 Joseph Marie Jacquard inventa el telar de Jaquard, el cual funciona a partir de un programa escrito de forma mecánica en un código.

1936 Máquina de Turing: Demostración de que hay problemas que una máquina no puede resolver.

1957 Se crea FORTRAN primer lenguaje de programación.

1969 El programa creado por el equipo a cargo de Margaret Hamilton "on-board" ayuda en el aterrizaje exitoso de la cápsula Eagle del Apollo 11.

1970 Se crea el lenguaje C

1989 Creación de Python.

# 1. SISTEMAS OPERATIVOS

![Imagen](https://www.aboutespanol.com/thmb/6fHKwkQXZIDNWu8NPr800Ev3cjc=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc():format(webp)/Que-Es-Sistema-Operativo-569cb26a5f9b58eba4aae8ae.png)

Lenguaje de programación: Conjunto de símbolos con estructura propia que permite al programador escribir instrucciones o secuencias de órdenes.

Lenguaje de máquina: Conjunto de símbolos con estructura propia a través de las cuales se puede controlar la unidad central de procesamiento CPU de un computador, pues son interpretables para la misma.

Algoritmos: Serie pasos a través de los cuales se puede realizar una tarea.

Programa: Código que se vale de uno o más algoritmos escrito en un lenguaje específico.
