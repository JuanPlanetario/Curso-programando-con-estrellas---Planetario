```python
n=0

while(n <= 5):
    print (n)
    n = n + 1
```

```python
n
```

```python
for n in range(5):
    print (n)
```

```python
for n in range(5):
    print (n)
    if(n > 2):
        print ("n es mayor a 2!!")
```
