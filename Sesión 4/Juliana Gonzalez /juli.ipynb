```python
plt.title("una recta linea")

plt.xlabel("eje x")
plt.ylabel("eje y")

plt.plot(x,recta(x,1,0), label="m=0, b=0")
plt.legend()
plt.axis("equal")
```

```python
plt.title("Varias bonitas rectas")

plt.xlabel("eje x")
plt.ylabel("eje y");

plt.plot( x, recta(x,1,0), label= "m=1, b=0")
plt.plot( x, recta(x,1,2), label= "m=0, b=2")
plt.plot( x, recta(x,1,4), label= "m=-2, b=4")


plt.legend()
plt.axis('equal')
```

```python
plt.title("Varias bonitas rectas")

plt.xlabel("eje x")
plt.ylabel("eje y");

plt.plot( x, recta(x,1,0), label= "m=1, b=0")
plt.plot( x, recta(x,0,0), label= "m=0, b=0")
plt.plot( x, recta(x,-2,0), label= "m=-2, b=0")


plt.legend() # esto me da etiquetas para saber quien es la recta
plt.axis('equal')
```

```python
fig = plt.figure(figsize=[7,7])
plt.xlim(-3,3) #dato inicil y final en x
plt.ylim(-6,6) #dato inicial y final en t

plt.title("Varias rectas muy bonitas")

plt.xlabel("eje x")
plt.ylabel("eje y");

plt.plot( x, recta(x,1,0), label= "m=1, b=0")
plt.plot( x, recta(x,0,0), label= "m=0, b=0")
plt.plot( x, recta(x,-2,0), label= "m=-2, b=0")

plt.legend()
plt.savefig("preguntita")
```

```python
def cuadratica(x,a,b,c):
     y= a*x**2 + b*x +c
     return y
```

```python
fig = plt.figure(figsize=[7,7])
plt.xlim(-3,3)
plt.ylim(-3,3)


plt.title("Una funci√≥n muy cuadr√°tica")

plt.xlabel("eje x")
plt.ylabel("eje y");

plt.plot( x, cuadratica(x,2,-1,-1), label= "$2 x^2 - x - 1$")
plt.plot( x, cuadratica(x,-2,1,1), label= "$2 x^2 - x - 1$")

plt.legend()
```
